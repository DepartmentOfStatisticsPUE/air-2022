Filtrujemy wiersze aby przyjrzeć się "dziwnym" wartościom

```python
rent[ rent.flat_area >= 500 ][ ["price", "flat_area", "ad_title"] ]
```

```python
rent[ rent.price >= 10000 ][ ["price", "flat_area", "ad_title"] ]
```

```python
rent[ rent.price <= 300][ ["price", "flat_area", "ad_title"] ]
```

```python
rent[(rent.price <= 500) & (rent.flat_area <= 20)]
```
