Model wykładniczy.

$$
y = e^{a+b \times x}
$$

Logarytm pozwala nam zlinearyzować model wykładniczy.

$$
\log(y) = \log(e^{a+b \times  x}) = a+b \times  x
$$

Tworzymy podsumowania zbioru według grup. Aby to zrobić musimy zastosować metodę `groupby` do grupowania ramki danych według jednej lub większej liczby zmiennych oraz metodę `agg` do zagregowania kolumn(y) według danej funkcji (np. średnia arytmetyczna).

building_type -- type of building
POL: 

* [0] blok
* [1] kamienica
* [2] dom wolnostojący
* [3] plomba
* [4] szeregowiec 
* [5] apartamentowiec 
* [6] loft


```python
rent.groupby("building_type").price.agg("mean")
```

```python
rent.groupby("building_type").price.mean() ## tylko jedna funkcja!!!
rent.groupby("building_type")["price"].mean() ## tylko jedna funkcja!!!
```

```python
rent.groupby("building_type", dropna=False).price.agg(["mean", "count"])
```

```python
rent.groupby("building_type", dropna=False).agg(srednia = pd.NamedAgg("price", "mean"),
                                                liczba = pd.NamedAgg("price", "count"))
```

```python
rent.groupby(["building_type", "flat_balcony"], dropna=False).price.agg(["median", "count"])
```

Teraz łączymy poszczególne kroki by ich wyniki przekazać dalej

```python
rent[["id", "price", "flat_area", "quarter"]].query(
    "price >= 1000 & price <= 10000 & flat_area <= 200").groupby(
        "quarter").agg(
            price_med = pd.NamedAgg("price", "median"),
            flat_med = pd.NamedAgg("flat_area", "median"),
            count = pd.NamedAgg("price", "count")).sort_values(
                by = "count", ascending = False)
```

```python
pd.set_option("max_rows", 100) ## zmienia liczbę wierszy, która jest pokazywana w wyniku
```
